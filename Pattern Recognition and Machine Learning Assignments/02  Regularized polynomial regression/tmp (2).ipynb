```python
def theta(x,y,lamd):
  identity=np.identity(x.shape[1])
  F=np.linalg.inv((x.T).dot(x)+lamd*identity).dot((x.T).dot(y))
  return F
```

```python
def RMSE(x,y,coeff):
  y_predicted=x.dot(coeff)
  rms=((np.sum((y_predicted-y)**2))/50)**0.5
  return [rms,y_predicted]
```
